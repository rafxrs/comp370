````python
df = pd.read_csv("data/tweets.csv")
df_english = df[df['language']=='English']
df_english_no_questions = df_english[df_english['content'].astype(str).str.contains(r"\?")]

df_english_no_questions.to_csv("data/tweets_english_no_q.csv", index=False) 

contents = df_english_no_questions['content']
contents.to_csv("data/contents.csv", index=False)

````

Data Annotation

````python
# find out whether our not the tweet mentions Trump with a bool trump_mention
# a tweet only mentions trump if it contains the word "Trump" case sensitive
# then we create a new csv with this new column
pattern = re.compile(r"(?<![A-Za-z0-9])Trump(?![A-Za-z0-9])")
df_english_no_questions['trump_mention'] = df_english_no_questions['content'].astype(str).apply(
    lambda s: pattern.search(s) is not None
)

df_english_no_questions.to_csv("data/tweets_en_no_q_trump.csv", index=False)
````

Data Analysis

````python
frac = len(df_english_no_questions[df_english_no_questions['trump_mention'] == True]) # only take the true ones
total = len(df_english_no_questions)
print(frac)
print(total)
print(frac / total)
````
